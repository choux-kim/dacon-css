## 2. Preprocessing [Scaling 하지 않음]

 - Index : 삭제
 - FLAG_MOBIL : 삭제
 - child_num : 삭제 (family size와의 상관성 0.89)
 - NaN값 : unknown으로 대체
 - One-Hot encoding : 범주형 변수 8개
 - Train / Valid / Test Spliting = 6 : 2 : 2
 - Robust Scaling : income_total, DAYS_EMPLOYED, and family_size (이상치 영향 제한)

```python
data_train = data_train.fillna('unknown')

X = data_train.drop(columns=['index', 'FLAG_MOBIL', 'child_num', 'credit'])
y = data_train['credit']

X_encoded = pd.get_dummies(X, columns=['gender', 'car', 'reality', 'income_type', 'edu_type', 'family_type', 'house_type', 'occyp_type'])
```

```python
X_encoded.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
list = [X_train, X_valid, X_test, y_train, y_valid, y_test]
for i in list:
  print(i.shape)
```

## 3. Support Vector Machine (kernel = Gaussian)

```python
svm_clf = SVC(kernel='rbf',
              probability = True,
              random_state=2021)
svm_clf.fit(X_train, y_train)
```

```python
y_pred_svm = svm_clf.predict_proba(X_valid)
```

## 4. Submission

```python
from datetime import datetime

curr_time = datetime.now().strftime('%y%m%d%H%M%S')
```

```python
data_test = data_test.fillna('unknown')
X_s = data_test.drop(columns=['index', 'FLAG_MOBIL', 'child_num'])
X_s_encoded = pd.get_dummies(X_s, columns=['gender', 'car', 'reality', 'income_type', 'edu_type', 'family_type', 'house_type', 'occyp_type'])
```

```python
l_idx = data_test['index']
submission = pd.DataFrame(np.insert(svm_clf.predict_proba(X_s_encoded), 0, l_idx, axis=1), columns=['index', 0, 1, 2])
submission['index'] = submission['index'].astype('int')
```

```python
submission.to_csv(f'submission_{curr_time}.csv', index=False)
```
