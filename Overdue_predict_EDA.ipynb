DAYS_EMPLOYED: 업무 시작일

```python
employed = - df["DAYS_EMPLOYED"].loc[df["DAYS_EMPLOYED"]<0]
unemployed = df["DAYS_EMPLOYED"].loc[df["DAYS_EMPLOYED"]>0]
```

```python
print('employed people:', len(employed))
print('unemployed people:', len(unemployed))
```

```python
plt.bar(np.arange(2), [len(employed), len(unemployed)]);
plt.xticks(np.arange(2), labels=['employed', 'unemployed']);
plt.title('status of employment');
```

```python
plt.hist(employed, bins=50);
plt.title("period of employment");
```

occyp_type: 직업 유형

```python
o_type = df['occyp_type'].value_counts();
plt.figure(figsize=(20,10))
ax = o_type.plot(kind='barh', title='type of occupation',rot=0)
for p in ax.patches:
    x, y, width, height = p.get_bbox().bounds
    ax.text(width*1.01, y+height/2, "%.0f "%(width), va='center')
```

begin_month: 신용카드 발급 월

```python
card = (-df["begin_month"]/12)
```

```python
plt.title('Usage period of credit card')
plt.hist(card, bins=30);
```

```python
card_2 = (- df["begin_month"])//12
card_2 = card_2.value_counts()
ax = card_2.plot(kind='bar', title='Usage period of credit card(only year)', rot=0)
for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("%.0f"%(height), (left+width/2, height*1.01), ha='center')
plt.show()
```
